```python
import pandas as pd
```

```python
csv_path='ds_team_project1/data/indices/ml_adjusted_ds_indices.csv'
o_df = pd.read_csv(csv_path)
new_all_df = o_df[['city','restaurant_price_index','cpi_index','rent_index','groceries_index']]
new_all_df.head()
```

```python
city_list = new_all_df['city'].tolist()
#city_list
```

```python
def ds(num):
  csv = city_list[num] + '_ds_indices.csv'
  csv_path = 'ds_team_project1/data/indices/nearby_cities/' + csv
  city_df = pd.read_csv(csv_path)
  new_df = city_df[['restaurant_price_index','cpi_index','rent_index', 'groceries_index']]
  new_list = new_df.append(new_all_df.loc[new_all_df['city'] == city_list[num]]).mean().tolist()
  df = pd.DataFrame({'city': city_list[num],
                  'restaurant_price_index':new_list[0],
                  'cpi_index':new_list[1],
                  'rent_index':new_list[2],
                  'groceries_index':new_list[3]},index=[num])
  return df
```

```python
# Use NY as Benchmark
```

```python
cols = ['city', 'cpi_index', 'groceries_index', 'rent_index', 'restaurant_price_index']
data = []
for x in range(42):
    data.append(ds(x).values.tolist()[0])
c_df = pd.DataFrame(data, columns=cols)
c_df.head()
```

```python
csv_path='ds_team_project1/data/salary/payscale_ds_salary.csv'
salary_df=pd.read_csv(csv_path)
salary_df.head()
```

```python
c_df['cpi_single']=c_df['cpi_index']/100*1957.34
c_df['cpi_family']=c_df['cpi_index']/100*5699.47
c_df['rent_single']=c_df['rent_index']/100*(3096.05+1886.53)/2
c_df['rent_family']=c_df['rent_index']/100*(6017.17+3321.3)/2
c_df['groceries_single']=c_df['groceries_index']/100*283.69
c_df['groceries_family']=c_df['groceries_index']/100*1418.43 
c_df['restaurant_single']=c_df['restaurant_price_index']/100*795.83 
c_df['restaurant_family']=c_df['restaurant_price_index']/100*2365.20
c_df['occupation']=salary_df['occupation']
c_df['salary']=salary_df['base_salary_md_pay']
c_df['col_single']= c_df['cpi_single'] + c_df['rent_single']
c_df['col_family']= c_df['cpi_family'] + c_df['rent_family']
c_df['ds_single']= salary_df['base_salary_md_pay']/12 - c_df['col_single']
c_df['ds_family']= salary_df['base_salary_md_pay']/12 - c_df['col_family']
```

```python
c_df['state'] = o_df['state']
new_c_df = c_df[['state','city','occupation','salary','cpi_index','rent_index','restaurant_price_index','groceries_index','cpi_single','cpi_family','rent_single','rent_family','groceries_single','groceries_family','restaurant_single','restaurant_family','col_single','col_family','ds_single','ds_family']]
new_c_df.head()
```

```python
top10_df=new_c_df.nlargest(10,'salary')
top10_df
```

```python
new_c_df.to_csv('ds_team_project1/data/indices/converted_indices_ny.csv')
top10_df.to_csv('ds_team_project1/data/indices/top10_indices_ny.csv')
```
