```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('../data/indices/converted_cities/converted_indices_la.csv')
```

```python
top10_df_single=df.nlargest(10,'ds_single')
top10_df_family=df.nlargest(10,'ds_family')
top10_df_family
```

```python
df.head(2)
```

```python
top10_df_single.plot(x="city", y="ds_single", kind="bar", figsize=(10,5),legend =False)
plt.xticks(rotation=60)
plt.title("Top Ten Most Profitable Cities for Data Scientist (single)")
plt.xlabel("City")
plt.ylabel("Monthly Profit ($)") 
plt.savefig("SingleDS.png")
plt.show()
```

```python
top10_df_family.plot(x="city", y="ds_family", kind="bar", figsize=(10,5),legend =False)
plt.xticks(rotation=60)
plt.title("Top Ten Most Profitable Cities for Data Scientist (family)")
plt.xlabel("City")
plt.ylabel("Monthly Profit ($)") 
plt.savefig("FamilyDS.png")
plt.show()
```
